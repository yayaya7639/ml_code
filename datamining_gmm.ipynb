```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
data = pd.read_csv("data_inakaortokai.csv",sep=",")
```

```python
data
```

```python
label = data["prefecture"]
label
```

```python
scaler = StandardScaler()
data_std = pd.DataFrame(scaler.fit_transform(data.drop(columns="prefecture")), columns=data.drop(columns="prefecture").columns)
```

```python
data_std
```

```python
data_dropbirth = data_std.drop(columns="number of births")
```

```python
from sklearn.decomposition import PCA

pca = PCA()
feature = pca.fit(data_dropbirth)
feature = pca.transform(data_dropbirth)
pca_data = pd.DataFrame(feature, columns=["PC{}".format(x + 1) for x in range(len(data_dropbirth.columns))])
```

```python
drop_data =  pca_data.drop(columns="PC3").drop(columns="PC4")
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
drop_data.plot(0,1,kind='scatter',ax=ax)

for k, v in drop_data.iterrows():
        ax.annotate(data["prefecture"][k],xy=(v[0],v[1]),size=10)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
from sklearn import mixture
gmm=mixture.GaussianMixture(n_components=2,covariance_type='full')
z_gmm=gmm.fit(drop_data)
z_gmm=z_gmm.predict(drop_data)

color = [0]*47
for i in range(47):
    if z_gmm[i] == 1:
        color[i] = 'r'
    else:
        color[i] = 'b'
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
drop_data.plot(0,1,kind='scatter',ax=ax, c=color)

for k, v in drop_data.iterrows():
        ax.annotate(data["prefecture"][k],xy=(v[0],v[1]),size=10)
        
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
